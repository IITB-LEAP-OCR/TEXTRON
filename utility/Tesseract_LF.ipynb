### Imports

```python
import os
import cv2
import glob
from pytesseract import Output, image_to_data
from skimage import io, color
from skimage.filters import threshold_otsu
```

### Get Bounding Boxes

```python
path = 'DocBank_Samples/data/*ori.jpg'
files = glob.glob(path)
print(files)
```

```python
output_dir = 'tesseract_images/'
```

```python
dimensions = {}
results = {}
pixels = {}
i =1
for file in files:
    values = []
    img = io.imread(file)
    d = image_to_data(img, output_type=Output.DICT)
    conf = d['conf']
    for i in range(len(d['level'])):
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        (x, y, w, h) = (int(x), int(y), int(w), int(h))
        values.append([conf[i], x, y, w, h])
        if(x==0 and y==0):
            continue
        else:
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255))
    dimensions[file] = values
    results[file] = d
    filename = file.split('/')[-1:][0]
    print("On Next..." + str(i))
    i = i + 1
    io.imsave(output_dir + filename, img)
```
