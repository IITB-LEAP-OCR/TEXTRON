```python
import json
f = open("annotation.json")
data = json.load(f)
data.keys()
```

```python
l1 = sorted(os.listdir("images"))
l2 = sorted([(i["file_name"],i["id"]) for i in data["images"]])
j=0
id_list = []
for image in l2:
    if j==len(l1):
        break
    if image[0]==l1[j]:
        id_list.append(image[1])
        j+=1
len(id_list)
```

```python
print(id_list[0])
data["images"][id_list[0]-1]
```

```python
data["annotations"][0]
```

```python
data["annotations"][1890000:1891000]
```

```python
img_to_ann = {i:[] for i in id_list}
140878 in img_to_ann.keys()
```

```python
for i,ann in enumerate(data["annotations"]):
    # print(ann[])
    if ann["image_id"] in img_to_ann.keys():
        img_to_ann[ann["image_id"]].append(ann["id"])
for k in img_to_ann:
    print(k,img_to_ann[k])
    break
```

```python
data["annotations"][1890659]
```

```python
# with open("10k_annotation.json","w") as f2:
res = {}
res["info"] = data["info"]
res["licenses"] = data["licenses"]
res["categories"] = data["categories"]
res["images"] = [data["images"][id-1] for i,id in enumerate(id_list)]
res["annotations"] = [data["annotations"][id-1] for k in img_to_ann for id in img_to_ann[k]]
```

```python
for i in range(len(res["images"])):
    res["images"][i].update({"id":i+1})
res["images"]
```

```python
td = {}
for i,id in enumerate(id_list):
    td[id]=i+1

td
```

```python
for i,ann in enumerate(res["annotations"]):
    res["annotations"][i].update({"id":i+1, "image_id": td[ann["image_id"]]})

res["annotations"]
```

```python
f2 = open("10k_annotations.json","w")
json.dump(res,f2)
f.close()
f2.close()
```
